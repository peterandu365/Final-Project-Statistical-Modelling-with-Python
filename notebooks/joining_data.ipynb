```python
# summary statistics:
# after the cleaning process, let's do another comparison between the two dataframes: the number of columns and rows, and the null portion for each column. Make a table to compare the two dataframes.

compare_stations_POIs_DF = pd.DataFrame(index=['FourSquare: DF_FS_stations_POIs', 'yelp: DF_yelp_stations_POIs'], columns=['# of columns', '# of rows', 'null portion'])
compare_stations_POIs_DF['# of columns'] = [DF_FS_stations_POIs.shape[1], DF_yelp_stations_POIs.shape[1]]
compare_stations_POIs_DF['# of rows'] = [DF_FS_stations_POIs.shape[0], DF_yelp_stations_POIs.shape[0]]
compare_stations_POIs_DF['null portion'] = [DF_FS_stations_POIs.isnull().sum().max()/DF_FS_stations_POIs.shape[0], DF_yelp_stations_POIs.isnull().sum().max()/DF_yelp_stations_POIs.shape[0]]
compare_stations_POIs_DF
```

```python
# Make copy of this dataframe to and put a 'win' or 'loose' string value for each cell, by comparing the upper and lower cells.
compare_stations_POIs_DF_readable = compare_stations_POIs_DF.copy()
compare_stations_POIs_DF_readable['# of columns'] = compare_stations_POIs_DF_readable['# of columns'].apply(lambda x: 'win' if x == compare_stations_POIs_DF_readable['# of columns'].max() else 'loose')
compare_stations_POIs_DF_readable['# of rows'] = compare_stations_POIs_DF_readable['# of rows'].apply(lambda x: 'win' if x == compare_stations_POIs_DF_readable['# of rows'].max() else 'loose')
compare_stations_POIs_DF_readable['null portion'] = compare_stations_POIs_DF_readable['null portion'].apply(lambda x: 'win' if x == compare_stations_POIs_DF_readable['null portion'].min() else 'loose')

# add a column to indicate the win / loose status, by calculating the times "win" occcurse and "loose" occurs, comparing the number of "win" for each row. If the number of "win"s is larger than the other row, give it a string "winner" / 'loose' right align the text in the cell. (x == 'win').sum() is prefered.
compare_stations_POIs_DF_readable['Overall'] = compare_stations_POIs_DF_readable.apply(lambda x: 'Win' if (x == 'win').sum() > 1 else 'Lose', axis=1)
compare_stations_POIs_DF_readable
```

```python
# Without POI_category_name, there are 2555 unique POIs; with POI_category_name, there are 2771 unique POIs. It means that there are 216 POIs with the same name but different categories.

# Therefore, we would create a new column to commedate the POI_category_name values in a set{} format. If there is only one category, then the set{} will have only one element. If there are more than one categories, then the set{} will have more than one elements.

DF_yelp_stations_POIs_pivot = DF_yelp_stations_POIs.pivot_table(index=['POI_yelp_id'], aggfunc={'station_id': 'count', 'station_has_ebikes': 'sum', 'station_renting': 'sum', 'station_returning': 'sum', 'station_empty_slots': 'sum', 'station_free_bikes': 'sum', 'station_normal_bikes': 'sum', 'station_ebikes': 'sum', 'station_slots': 'sum'}).reset_index() # , 'POI_distance': 'median', 'POI_distance': 'min', 'POI_distance': 'max', 'POI_distance': 'std' })
DF_yelp_stations_POIs_pivot.shape
```

```python
# from DF_yelp_stations_POIs, groupby POI_yelp_id, and then aggregate the POI_distance 's min max mean median std into new columns
distance_min = DF_yelp_stations_POIs.pivot_table(index=['POI_yelp_id'], aggfunc={'POI_distance': 'min'}).reset_index()
# rename the column name
distance_min.rename(columns={'POI_distance': 'POI_distance_min'}, inplace=True)
distance_min.shape
distance_max = DF_yelp_stations_POIs.pivot_table(index=['POI_yelp_id'], aggfunc={'POI_distance': 'max'}).reset_index()
distance_max.rename(columns={'POI_distance': 'POI_distance_max'}, inplace=True)
distance_max.shape
distance_mean = DF_yelp_stations_POIs.pivot_table(index=['POI_yelp_id'], aggfunc={'POI_distance': 'mean'}).reset_index()
distance_mean.rename(columns={'POI_distance': 'POI_distance_mean'}, inplace=True)
distance_mean.shape
distance_median = DF_yelp_stations_POIs.pivot_table(index=['POI_yelp_id'], aggfunc={'POI_distance': 'median'}).reset_index()
distance_median.rename(columns={'POI_distance': 'POI_distance_median'}, inplace=True)
distance_median.shape

# distance_std = DF_yelp_stations_POIs.pivot_table(index=['POI_yelp_id'], aggfunc={'POI_distance': 'std'}).reset_index()
# distance_std.rename(columns={'POI_distance': 'POI_distance_std'}, inplace=True)
# distance_std.shape
# shape is (2371, 2), not as expected (2555, 2)

# distance_25_percentile, the 25 percentile of the POI_distance
distance_25_percentile = DF_yelp_stations_POIs.pivot_table(index=['POI_yelp_id'], aggfunc={'POI_distance': lambda x: np.percentile(x, 25)}).reset_index()
distance_25_percentile.rename(columns={'POI_distance': 'POI_distance_25_percentile'}, inplace=True)
distance_25_percentile.shape

# distance_75_percentile, the 75 percentile of the POI_distance
distance_75_percentile = DF_yelp_stations_POIs.pivot_table(index=['POI_yelp_id'], aggfunc={'POI_distance': lambda x: np.percentile(x, 75)}).reset_index()
distance_75_percentile.rename(columns={'POI_distance': 'POI_distance_75_percentile'}, inplace=True)
distance_75_percentile.shape

# distance_IQR, the IQR of the POI_distance
distance_IQR = DF_yelp_stations_POIs.pivot_table(index=['POI_yelp_id'], aggfunc={'POI_distance': lambda x: np.percentile(x, 75) - np.percentile(x, 25)}).reset_index()
distance_IQR.rename(columns={'POI_distance': 'POI_distance_IQR'}, inplace=True)
distance_IQR.shape
```
